```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None
        
class HashTable:
    def __init__(self, capacity):
        self.capacity = capacity
        self.size = 0
        self.buckets = [None] * self.capacity
        
    def put(self, key, value):
        my_hash = self.get_index(key)
        head = self.buckets[self.get_index(key)]
        if not head:
            self.size += 1
            self.buckets[self.get_index(key)] = Node(key, value)
            print(f'ключ {key} успешно добавлен со значением {value} и с хэш значением {my_hash}')
            return 
        prev = head
        while head:
            if key == head.key:
                head.value = value
                print(f'ключ {key} успешно добавлен со значением {value} и с хэш значением {my_hash}')
                return
            prev = head
            head = head.next
        prev.next = Node(key, value)
        self.size += 1
        print(f'ключ {key} успешно добавлен со значением {value} и с хэш значением {my_hash}')

    def get_index(self, key):
        my_hash_key = 0
        for el in key:
            my_hash_key += ord(el)
        return my_hash_key % self.capacity
        
    def find(self, key):
        node = self.buckets[self.get_index(key)]
        while node is not None and node.key != key:
            node = node.next
        if node is None:
            print('такого элемента нет')
            return None
        else:
            print(f'вот ваш элемент {node.value}')
            return node.value
    
    def remove(self, key):
        node = self.buckets[self.get_index(key)]
        prev = None
        while node is not None and node.key != key:
            prev = node
            node = node.next
        if node is None:
            print(f'ключа {key} нету в таблице')
            return 
        else:
            self.size -= 1
            result = node.value
            if prev is None:
                self.buckets[self.get_index(key)] = None
            else:
                perv.next = prev.next.next
            print(f'ключ {key} успешно удален со значением {result}')
            return result 
        
    def fill_rate(self):
        count = 0
        for bucket in self.buckets:
            if bucket is not None:
                count += 1
        return count/self.capacity
    
    def chain_length_calculation(self):
        list_lens = []
        for slot in self.buckets:
            if slot is not None:
                pass
            count = self.height(slot)
            list_lens.append(count)
        return max(list_lens), min([i for i in list_lens if i != 0])
    
    def height(self, slot):
        if not slot:
            return 0
        leng = self.height(slot.next)
        return  leng + 1
            
```

# 2 Варинат (Хэш таблица открытой адресации для разрешение коллизий)

```python
class Hash_op_adr:

    def __init__(self, table_size):
        self.hash_table = {}
        self.table_size = table_size
        
    def create_table(self):
        i = 0
        while i < self.table_size:
            self.hash_table[i] = (None)
            i += 1
            
    def print_table(self):
        for k, v in self.hash_table.items():
            print(k, end=' --> ')
            print(v, sep=' -- ', end = '\n')
            print()

    def hashing(self, key):
        return key%self.table_size
              
    def insert(self, value):
        hash_key = self.hashing(value[0])

        while self.hash_table[hash_key] != None:
            if hash_key == self.table_size-1:
                hash_key = -1
                
            hash_key += 1
            
        self.hash_table[hash_key] = value

            
    def delete(self, key):
        hash_key = self.hashing(key)
        while self.hash_table[hash_key][0] != key:
            hash_key += 1
        self.hash_table[hash_key] = (None)
        
    def find(self, key):
        hash_key = self.hashing(key)
        count = 0
        while self.hash_table[hash_key][0] != key:
            hash_key += 1
        print(f'{hash_key} --> {self.hash_table[hash_key]}')
```

```python
def gen_value():
    alp = [chr(el) for el in range(32, 250)]
    res = ''
    for _ in range(random.randint(5, 30)):
        res += random.choice(alp)
    return res
```

```python
table1 = Hash_op_adr(10000)
table1.create_table()
table1.print_table()
```

```python
for key in set([random.randint(1,1000000) for el in range(9000)]):
    val = gen_value()
    table1.insert((key, val))
table1.print_table()
```
